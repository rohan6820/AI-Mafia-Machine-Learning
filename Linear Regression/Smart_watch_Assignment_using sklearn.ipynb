```python
import numpy as n
import pandas as p
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df_x=p.read_csv("Linear_X_Train.csv").values
df_y=p.read_csv("Linear_Y_Train.csv").values
```

```python
plt.scatter(df_x,df_y)
plt.show()
```

```python
r=LinearRegression()
```

```python
r.fit(df_x,df_y)
```

```python
predicted=r.predict(df_x)
```

```python
plt.plot(df_x,predicted,color="red")
plt.scatter(df_x,df_y,color="lightblue")
plt.show()
```

```python
x_test=p.read_csv("Linear_X_Test.csv")
```

```python
y_pred=r.predict(x_test)
```

```python
plt.plot(x_test,y_pred,color="Red")
plt.scatter(df_x,predicted,color="lightblue")
plt.show()
```

```python
x_test["y_pred"]=y_pred
```
